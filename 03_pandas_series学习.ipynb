```python
# 案例练习
'''
1. 学生成绩统计：创建一个包含10名学生数学成绩的Series，成绩范围在50-100之间。计算平均分、最高分、最低分，并找出高于平均分的学生人数。
'''
import pandas as pd
import numpy as np

np.random.seed(42)
values = np.random.randint(50, 101, 10)
index = ['学生' + str(i) for i in range(1, 11)]
scores = pd.Series(values, index)
print(scores)

print(f'平均分：{scores.mean()}')
print(f'最高分：{scores.max()}')
print(f'最低分：{scores.min()}')
print(f'高于平均分的人数：{scores[scores > scores.mean()].count()}')
```

```python
'''
2. 温度数据分析：给定某城市一周每天的最高温度Series，完成任务：找出温度超过30度的天数，计算平均温度，将温度从高到低排序，找出温度变化最大的两天。
'''
temperatures = pd.Series([28, 31, 29, 32, 30, 27, 33],
                         index=['周一', '周二', '周三', '周四', '周五', '周六', '周日'])
print(f'超过30度的天数：{temperatures[temperatures > 30].count()}天')
print(f'平均温度：{temperatures.mean()}度')
print(f'温度排序：\n{temperatures.sort_values(ascending=False)}')
# diff(): 计算series数组的变化值
dif = temperatures.diff().abs()
# nlargest(n): 返回最大的n个元素
max_dif_days = dif.nlargest(2).index.tolist()
print(f'温度变化最大的两天：{max_dif_days}')
```

```python
'''
3. 股票价格分析：给定某股票连续10个交易日的收盘价Series：计算每日收益率（当日收盘价/前日收盘价 - 1），找出收益率最高和最低的日期，计算波动率（收益率的标准差）。
'''
prices = pd.Series([102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2],
                   index=pd.date_range('2026-02-01', periods=10))

# 计算每日收益率，pct:
pct = prices.pct_change()
print(f'每日收益率：\n{pct}')
# 收益率最高的日期
print(f'收益率最高的日期：{pct.idxmax()}')
print(f'收益率最低的日期：{pct.idxmin()}')
print(f'波动率（收益率的标准差）:{pct.std()}')
```

```python
'''
4. 销售数据分析：某产品过去12个月的销售量Series：计算季度平均销量（每3个月为一个季度），找出销量最高的月份，计算月环比增长率，找出连续增长超过2个月的月份
常用频率别名：月初 MS，月末 ME，日 D，周 W，季度 Q，年 A，小时 h，分钟 T
'''
sales = pd.Series([120, 135, 145, 160, 155, 170, 180, 175, 190, 200, 210, 220],
                  index=pd.date_range('2026-01-01', periods=12, freq='MS'))
# 季度的平均销量 resample: 对时间序列数据重新采样
# print(f'季度平均销量：\n{sales.resample('QS').mean()}')
# print(f'销量最高的月份：{sales.idxmax()}')
# print(f'月环比的增长率：\n{sales.pct_change()}')

pct = sales.pct_change()
pct2 = pct > 0
pct3 = pct2.rolling(3).sum() == 3
print(f'连续增长超过2个月的月份：{pct3[pct3 == True].keys().tolist()}')
```

```python
'''
5. 每小时销售数据分析：某商店每小时销售额Series：按天重采样计算每日总销售额，计算每天营业时间（8点-22点）和非营业时间的销售额比例，找出销售额最高的3个小时
'''
np.random.seed(42)
hourly_sales = pd.Series(np.random.randint(0, 100, 24), index=pd.date_range('2026-01-01', freq='h', periods=24))

day_sales = hourly_sales.resample('D').sum()
print(f'按天重采样计算每日总销售额：\n{day_sales}')

# 筛选一段时间内的series
# hourly_sales.between_time('8:00', '22:00')
# 营业时间的销售额
business_hour_sales = hourly_sales[(hourly_sales.index.hour >= 8) & (hourly_sales.index.hour <= 22)]
# 非营业时间的销售额 drop(): 根据索引去除部分元素
not_business_hour_sales = hourly_sales.drop(business_hour_sales.index)
print(f'营业时间和非营业时间销售额比例：{business_hour_sales.sum() / not_business_hour_sales.sum()}')
# print(f'营业时间和非营业时间销售额比例：{business_hour_sales.sum() / (day_sales - business_hour_sales.sum())}')
print(f'销售额最高的3个小时：\n{hourly_sales.nlargest(3)}')
```
