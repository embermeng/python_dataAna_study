```python
import pandas as pd
import numpy as np

# dataframe的创建方式
# 通过series创建
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([6, 7, 8, 9, 10])
df = pd.DataFrame({
    '第一列': s1,
    '第二列': s2
})
# print(type(df))
# print(type(df['第一列']))
# 通过字典创建
df = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'name': ['唐僧', '孙悟空', '猪八戒', '沙和尚', '白龙马'],
    'age': [30, 999, 888, 777, 666],
    'score': [66.6, 77.7, 88.8, 99.9, 100]
}, index=[1, 2, 3, 4, 5], columns=['name', 'score', 'age'])
df
```

```python
# dataframe的属性
print(f'行索引：{df.index}')
print(f'列标签：{df.columns}')
print(f'值：{df.values}')
```

```python
print(f'维度：{df.ndim}')
print(f'数据类型：\n{df.dtypes}')
print(f'形状：{df.shape}')
print(f'元素个数：{df.size}')
```

```python
# 转置
# print(df.T)
df
```

```python
# 获取元素
# 某行
print(df.loc[1])
print(df.iloc[0])
```

```python
# 某列
print(df.loc[:, 'name'])
print(df.iloc[:, 0])
```

```python
# 单个数据
print(df.at[3, 'score'])
print(df.iat[2, 1])
print(df.loc[3, 'score'])
print(df.iloc[2, 1])
```

```python
# 获取单列数据
print(df['name'])
print(df.name)
```

```python
# 获取多列数据
df[['name', 'score']]
```

```python
# 查看部分数据
df.head(2)
```

```python
df.tail(3)
```

```python
# 使用布尔索引筛选数据
print(df[df.score > 80])
df[(df.score > 80) & (df.age < 800)]
```
