```python
# 缺失值的处理
s = pd.Series([1, 2, np.nan, None, pd.NA])
df = pd.DataFrame([[1, pd.NA, 2], [2, 3, 5], [pd.NA, 4, 6]], columns=['第一列', '第二列', '第三列'])
# print(s)
# 查看是否是缺失值
print(s.isna())
print(s.isnull())
print(df.isna())
print(df.isnull())
# 查看缺失值的个数
print(s.isna().sum())
print(df.isna().sum())
```

```python
# 剔除缺失值
print(s.dropna())
# 只要一条数据里有缺失值，就删除这条数据
print(df.dropna())
# 如果所有的值都是缺失值，才删除这条数据
print(df.dropna(how='all'))
# 至少有n个值不是缺失值，就保留
print(df.dropna(thresh=2))
# 删除一整列的记录
print(df.dropna(axis=1))
# 如果某列有缺失值，则删除这一行
print(df.dropna(subset=['第一列']))
```

```python
# 填充缺失值
df = pd.read_csv('./data/weather_withna.csv')
df.tail()
# df.isna().sum(axis=0)
# 对缺失值填充
df.fillna({'temp_max': 20, 'wind': 2.5}).tail()
# 使用统计值填充
df.fillna(df[['temp_max', 'wind']].mean()).tail()
# 用前面的值填充，用后面的值填充: bfill()
df.ffill().tail()
```

```python
# 数据分箱
df = pd.read_csv('./data/employees.csv')
df1 = df.head(10)[['employee_id', 'salary']]
print(df1)
# bins=n：分成n段区间，起始值、结束值是所有数据的最小值、最大值
pd.cut(df1.salary, bins=2).value_counts()
# 自定义区间
df1['收入范围'] = pd.cut(df1.salary, bins=[0, 10000, 15000, 20000, 30000], labels=['低', '中', '高', '超高'])
# print(df1)
# 等频率分割
pd.qcut(df1.salary, 3).value_counts()
```

```python
# 分组聚合
# df.groupby('分组的字段')['聚合的字段'].聚合函数()
df = pd.read_csv('./data/employees.csv')
df = df.dropna(subset=['department_id'])
df['department_id'] = df['department_id'].astype('int64')
# 计算不同部门的平均薪资
df.groupby('department_id').groups  # 查看分组情况
df.groupby('department_id').get_group(20)  # 查看具体的某个分组数据
df2 = df.groupby('department_id')[['salary']].mean().round(2)
df2 = df2.reset_index()
df2.sort_values('salary', ascending=False)
```

```python
# 计算不同部门不同岗位的人的平均薪资
df2 = df.groupby(['department_id', 'job_id'])['salary'].mean()
df2 = df2.reset_index()
df2['salary'] = df2['salary'].round(1)
df2.sort_values('salary', ascending=False)
```
