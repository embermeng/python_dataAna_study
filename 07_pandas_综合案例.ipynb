```python
# 企鹅数据分析
# 1. 导入必要的库
import numpy as np
import pandas as pd

# 2. 导入数据
df = pd.read_csv('./data/penguins.csv')
```

```python
# 3. 数据清洗
# 处理缺失值
df.dropna(inplace=True)  # inplace=True: 修改原数据
```

```python
# 4. 数据特征的构造
df['sex'] = df['sex'].astype('category')
df['bill_radio'] = df['bill_length_mm'] / df['bill_depth_mm']
df.head(5)
```

```python
# 5. 数据分析
# 数据分箱：把体重分为三个等级
df['mass_level'] = pd.cut(df['body_mass_g'], bins=3, labels=['低', '中', '高'])
df['mass_level'].value_counts()
# 按岛屿、性别分组分析
# df.groupby(['sex'])['body_mass_g'].mean()
# df.groupby(['sex'])['body_mass_g'].count()
df.groupby(['sex', 'island']).agg({
    'body_mass_g': ['mean', 'count']
})
```

```python
# 睡眠质量分析
import pandas as pd

df = pd.read_csv('./data/sleep.csv')
df.head(5)

# df.isna().sum()
# df['sleep_disorder'].value_counts()
df.drop(columns='sleep_disorder', inplace=True)

df['gender'] = df['gender'].astype('category')
df['occupation'] = df['occupation'].astype('category')
df['bmi_category'] = df['bmi_category'].astype('category')
df[['high', 'low']] = df['blood_pressure'].str.split('/', expand=True)
df.head(2)

# 睡眠质量的分箱
df['quality_level'] = pd.cut(df['sleep_quality'], bins=3, labels=['差', '中', '优'])
df['age_level'] = pd.cut(df['age'], bins=3, labels=['青少年', '中年', '老年'])
df.head(2)

# df['bmi_category'].value_counts()
# 根据不同的bmi分组，看睡眠质量是否有区别
df.groupby(['age_level', 'bmi_category']).agg({
    'sleep_duration': ['mean'],
    'sleep_quality': ['mean'],
    'stress_level': ['mean'],
})
```
