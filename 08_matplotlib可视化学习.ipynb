```python
# 折线图
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['font.family'] = 'SimHei'  # 设置中文字体

# 创建图表，设置大小
plt.figure(figsize=(10, 5))

month = ['1月', '2月', '3月', '4月']
sales = [100, 150, 80, 130]

# 绘制
plt.plot(month, sales, label='产品A', color='orange', linewidth=2, linestyle='--', marker='o')
plt.title('2025年销售趋势', color='#fff', fontsize=20)
plt.xlabel('月份', fontsize=10)
plt.ylabel('销售额（万元）', fontsize=10)
plt.legend(loc='upper left')
# 网格线
plt.grid(True, alpha=0.2, color='#bfa', linestyle='--')
# 刻度字体大小
plt.xticks(fontsize=12, rotation=0)
plt.yticks(fontsize=12, rotation=0)
# y轴范围
plt.ylim(0, 160)
# 每个数据上显示数值
for x, y in zip(month, sales):
    print(x, y)
    plt.text(x, y + 1, str(y), ha='center', va='bottom', fontsize=10)

# 显示图表
plt.show()
```

```python
# 柱状图
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['font.family'] = 'SimHei'  # 设置中文字体

# 创建图表，设置大小
plt.figure(figsize=(10, 5))

sub = ['语文', '数学', '英语', '物理', '化学', '生物']
scores = [92, 88, 99, 60, 50, 56]
# 绘制
plt.bar(sub, scores, label='小红', color='#bfa')
plt.title('小红2025年成绩分布', color='#fff', fontsize=20)
plt.xlabel('科目', fontsize=10)
plt.ylabel('分数', fontsize=10)
plt.legend(loc='upper right')
# 网格线 只要横向
plt.grid(axis='y', alpha=0.2, color='#bfa', linestyle='--')
# y轴范围
plt.ylim(0, 120)
# 每个数据上显示数值
for x, y in zip(sub, scores):
    plt.text(x, y + 1, str(y), ha='center', va='bottom', fontsize=10)

# 自动优化排版
# plt.tight_layout()
# 显示图表
plt.show()
```

```python
# 条形图
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['font.family'] = 'SimHei'  # 设置中文字体

# 创建图表，设置大小
plt.figure(figsize=(10, 5))

sub = ['语文', '数学', '英语', '物理', '化学', '生物']
scores = [92, 88, 99, 60, 50, 56]
# 绘制
plt.barh(sub, scores)
# 显示图表
plt.show()
```

```python
# 饼图
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['font.family'] = 'SimHei'  # 设置中文字体

# 创建图表，设置大小
plt.figure(figsize=(10, 5))

todos = ['学习', '娱乐', '运动', '睡觉', '其他']
times = [6, 4, 1, 8, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499']
# 绘制
plt.pie(times, labels=todos,
        autopct='%.1f%%',  # 显示百分比
        startangle=90,  # 调整初始画图的角度
        colors=colors,
        wedgeprops={'width': 0.3},  # 设置圆环
        pctdistance=0.85,  # 环中文字距离圆心的距离
        )
plt.title('时间分布', color='#fff', fontsize=20)
# 中心文字
plt.text(0, 0, '总计：24h', ha='center', va='center', fontsize=10)
plt.legend(loc='upper right')

# 显示图表
plt.show()
```
