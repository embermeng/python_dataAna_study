```python
import pandas as pd
import numpy as np
```

```python
'''
案例1：学生成绩分析
场景：某班级的学生成绩数据如下，请完成以下任务：
1. 计算每位学生的总分和平均分。
2. 找出数学成绩高于90分或英语成绩高于85分的学生。
3. 按总分从高到低排序，并输出前3名学生。
'''
data = {
    '姓名': ['张三', '李四', '王五', '赵六', '钱七'],
    '数学': [85, 92, 78, 88, 95],
    '英语': [90, 88, 85, 92, 80],
    '物理': [75, 80, 88, 85, 90]
}
scores = pd.DataFrame(data)
# sum()默认是竖着求和，如果想横着求和需传axis=1
scores['总分'] = scores[['数学', '英语', '物理']].sum(axis=1)
scores['平均分'] = scores['总分'] / 3
scores['平均分2'] = scores[['数学', '英语', '物理']].mean(axis=1)
highScores = scores[(scores['数学'] > 90) | (scores['英语'] > 85)]
# scores.nlargest(3, columns=['总分'])
top3 = scores.sort_values('总分', ascending=False).head(3)

print(f'总分和平均分：\n{scores}', )
print(f'数学成绩高于90分或英语成绩高于85分的学生：\n{highScores}')
print(f'总分前3：\n{top3}')
```

```python
'''
案例2：销售数据分析
场景：某公司销售数据如下，请完成以下任务：
1. 计算每种产品的总销售额（销售额 = 单价 × 销量）。
2. 找出销售额最高的产品。
3. 按销售额从高到低排序，并输出所有产品信息。
'''
data = {
    '产品名称': ['A', 'B', 'C', 'D'],
    '单价': [100, 150, 200, 120],
    '销量': [50, 30, 20, 40]
}
df = pd.DataFrame(data)
df['销售额'] = df['单价'] * df['销量']
sortedDf = df.sort_values('销售额', ascending=False)
maxDf = sortedDf.head(1)

print(f'总销售额：\n{df}')
print(f'销售额最高的产品：\n{maxDf}')
print(f'销售额从高到低排序：\n{sortedDf}')
```

```python
'''
案例3：员工考勤统计
场景：某部门员工考勤数据如下，请完成以下任务：
1. 计算每位员工的出勤率（出勤率 = 出勤天数 / 工作日总数）。
2. 标记出勤率低于80%的员工。
3. 按出勤率从高到低排序。
'''
data = {
    '姓名': ['张三', '李四', '王五', '赵六'],
    '出勤天数': [20, 15, 18, 22],
    '工作日总数': [25, 20, 25, 25]
}
df = pd.DataFrame(data)
df['出勤率'] = df['出勤天数'] / df['工作日总数']
lowDf = df[df['出勤率'] < 0.8]
sortedDf = df.sort_values('出勤率', ascending=False)

print(f'出勤率：\n{df}')
print(f'出勤率低于80%的员工：\n{lowDf}')
print(f'出勤率从高到低排序：\n{sortedDf}')
```

```python
'''
案例4：电影评分分析
场景：某电影评分数据如下，请完成以下任务：
1. 计算每部电影的平均评分。
2. 找出评分高于8.5的电影。
3. 按平均评分从高到低排序。
'''
data = {
    '电影名称': ['电影A', '电影B', '电影C', '电影D'],
    '评分1': [9.0, 8.5, 8.0, 7.5],
    '评分2': [8.5, 9.0, 8.5, 8.0],
    '评分3': [9.5, 8.0, 7.5, 7.0]
}
df = pd.DataFrame(data)
df['平均评分'] = df[['评分1', '评分2', '评分3']].mean(axis=1)
highDf = df[df['平均评分'] > 8.5]
sortedDf = df.sort_values('平均评分', ascending=False)

print(f'平均评分：\n{df}')
print(f'评分高于8.5的电影：\n{highDf}')
print(f'平均评分从高到低排序：\n{sortedDf}')
```

```python
'''
案例5：股票价格分析
场景：某股票价格数据如下，请完成以下任务：
1. 计算每日股价的涨跌幅（涨跌幅 = (当日收盘价 - 前一日收盘价) / 前一日收盘价）。
2. 找出涨幅超过5%的日期。
3. 按日期排序，并输出涨跌幅最高的日期。
'''
data = {
    '日期': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04'],
    '收盘价': [100, 105, 110, 102]
}
df = pd.DataFrame(data)
# 将字符串日期转换成日期时间类型，方便排序
df['日期'] = pd.to_datetime(df['日期'])
df['涨跌幅'] = df['收盘价'].pct_change()
highDf = df[df['涨跌幅'] > 0.05]
maxDate = df.sort_values('日期').nlargest(1, '涨跌幅')['日期']

print(f'每日股价的涨跌幅:\n{df}')
print(f'涨幅超过5%的日期:\n{highDf}')
print(f'涨跌幅最高的日期:\n{maxDate}')
```

```python
'''
案例 6：电商用户行为分析（基础版）
场景：某电商平台的用户行为数据如下，请完成以下任务：
1. 计算每位用户的总消费金额（消费金额 = 商品单价 × 购买数量）
2. 找出消费金额最高的用户，并输出其所有信息
3. 计算所有用户的平均消费金额（保留2位小数）
4. 统计电子产品的总购买数量
'''
data = {
    '用户ID': [101, 102, 103, 104, 105],
    '用户名': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    '商品类别': ['电子产品', '服饰', '电子产品', '家居', '服饰'],
    '商品单价': [1200, 300, 800, 150, 200],
    '购买数量': [1, 3, 2, 5, 4]
}
df = pd.DataFrame(data)
df['消费金额'] = df['商品单价'] * df['购买数量']
maxDf = df[df['消费金额'] == df['消费金额'].max()]
avg = round(df['消费金额'].mean(), 2)
total = df[df['商品类别'] == '电子产品']['购买数量'].sum()

print(f'总消费金额:\n{df}')
print(f'消费金额最高的用户:\n{maxDf}')
print(f'平均消费金额:{avg}')
print(f'总购买数量:{total}')
```
